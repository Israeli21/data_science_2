```python
# Including Python Libraries
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt("data.txt", delimiter="\t")
```

```python
# Part 1: Preparing the Data
# Read the data into X (features) and y (labels) variables
X = data[:, 0:2]
y = data[:, 2]

# Print the shape of X and y
print(f"Shape of X: {X.shape}")
print(f"Shape of y: {y.shape}")

# Shuffle the data and create train/test splits: X_train, X_test, y_train, y_test
np.random.seed(42)
indices = np.random.permutation(len(X))
test_size = int(0.3 * len(X))
test_indices = indices[:test_size]
train_indices = indices[test_size:]

X_train = X[train_indices]
X_test = X[test_indices]
y_train = y[train_indices]
y_test = y[test_indices]

# Normalize the data (mean zero, unit varience)
train_mean = np.mean(X_train, axis=0)
train_std = np.std(X_train, axis=0)

X_train_scaled = (X_train - train_mean) / train_std
X_test_scaled = (X_test - train_mean) / train_std

# Verify the scaling worked
print("X_train scaled mean:", np.mean(X_train_scaled, axis=0))
print("X_train scaled std:", np.std(X_train_scaled, axis=0))
```

```python
# Part 2: Plotting the Data
# Create a scatter plot of the training data to visualize
import matplotlib.patches as mpatches

plt.figure(figsize=(10, 4))
plt.scatter(X_train_scaled[:, 0], X_train_scaled[:, 1], c=y_train, cmap='bwr', alpha=0.7, edgecolors='k')
plt.title("Training Set")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
legend_handles = [
    mpatches.Patch(color='blue', label='label 0'),
    mpatches.Patch(color='red', label='label 1')
]
plt.legend(handles=legend_handles)
plt.show()

# Create a scatter plot of the test data to visualize.
plt.figure(figsize=(10, 4))
plt.scatter(X_test_scaled[:, 0], X_test_scaled[:, 1], c=y_test, cmap='bwr', alpha=0.7, edgecolors='k')
plt.title("Test Set")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")

plt.legend(handles=legend_handles)
plt.show()
```

```python
# Part 3: Defining the Perceptron Model
# Complete the code of the Perceptron model based on what we've provided in the next slide
class Perceptron():
    def __init__(self, num_features, learning_rate = 0.1):
        self.num_features = num_features
        self.lr = learning_rate
        self.weights = np.zeros(num_features)  # Initialize weights to 0
        self.bias = 0.0                        # Initilize bias to 0

    def forward(self, x):
        linear = np.dot(x, self.weights) + self.bias
        predictions = np.where(linear >= 0, 1, 0)
        return predictions

    def backward(self, x, y):
        predictions = self.forward(x)
        error = y - predictions

        # Update weights and bias using perceptron rule
        for i in range(x.shape[0]):
            self.weights += self.lr * errors[i] * x[i]
            self.bias += self.lr * errors[i]
        return errors

    def train(self, x, y, epochs):
        for e in range(epochs):
            for i in range(y.shape[0]):
                prediction = self.forward(x[i])
                error = y[i] - prediction

                # weight and bias update (perceptron rule)
                self.weights += self.lr * error * x[i]
                self.bias += self.lr * error

    def evaluate(self, x, y):
        predictions = self.forward(x)
        accuracy = np.mean(predictions == y)
        return accuracy
```

```python
# Part 4: Training the Perceptron
# Using your class and data from the previous slides, create and train the perceptron for 5 epochs
perceptron = Perceptron(num_features=2, learning_rate=0.1)
perceptron.train(X_train_scaled, y_train, epochs=5)
acc = perceptron.evaluate(X_test_scaled, y_test)
print(f"Test Accuracy after 5 epochs: {acc:.2f}")
```

```python
# Part 5: Evaluating the Model
# Compute the prediction accuracy (in percent) for both the training set and the test set.
# We get around 100% train set accuracy and 93% test set accuracy
train_acc = perceptron.evaluate(X_train_scaled, y_train) * 100
test_acc = perceptron.evaluate(X_test_scaled, y_test) * 100

print(f"Training Accuracy: {train_acc:.2f}%")
print(f"Test Accuracy: {test_acc:.2f}%")
```

```python
# Part 6: Decision Boundary
# Plot the 2d Decision Boundary for your train set
# Plot the 2d Decision Boundary for your test set
def plot_decision_boundary(perceptron, X, y, title):
    # Scatter plot of points
    plt.figure(figsize=(8, 6))
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap='bwr', alpha=0.7, edgecolors='k')

    # Get weights and bias
    w1, w2 = perceptron.weights
    b = perceptron.bias

    # Generate x values (span the feature space)
    x_min, x_max = X[:,0].min() - 1, X[:,0].max() + 1
    x_vals = np.linspace(x_min, x_max, 100)

    # Compute decision boundary line
    if w2 != 0:  # avoid division by zero
        y_vals = -(w1/w2) * x_vals - (b/w2)
        plt.plot(x_vals, y_vals, 'k--', linewidth=2, label="Decision Boundary")

    plt.title(title)
    plt.xlabel("Feature 1 (scaled)")
    plt.ylabel("Feature 2 (scaled)")
    plt.legend()
    plt.show()

# Plot for training set
plot_decision_boundary(perceptron, X_train_scaled, y_train, "Decision Boundary - Training Set")

# Plot for test set
plot_decision_boundary(perceptron, X_test_scaled, y_test, "Decision Boundary - Test Set")
```
